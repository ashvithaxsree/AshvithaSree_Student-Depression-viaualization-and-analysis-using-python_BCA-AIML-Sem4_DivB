```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/student_depression_dataset.csv')

# Display first few rows
df.head()
```

```python
import matplotlib.pyplot as plt

gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
# Replace 'Depression Level' with the actual column name in your dataset
sns.countplot(x='Depression', data=df, palette='pastel')
plt.title('Student Count by Depression Level')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Students')
plt.xlabel('Age')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.figure(figsize=(8,5))
# Replace 'Depression Level' with the actual column name for depression level in your dataset, for example, 'Depression'
sns.boxplot(x='Depression', y='Age', data=df, palette='coolwarm')
plt.title('Age Distribution Across Depression Levels')
plt.xticks(rotation=45)
plt.show()
```

```python
correlation = df.corr(numeric_only=True)
plt.figure(figsize=(10,6))
sns.heatmap(correlation, annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Countplot for academic pressure
sns.countplot(data=df, x='academic_pressure', palette='magma')
plt.title('Academic Pressure among Students')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count of students in each depression category
sns.countplot(x='depression', data=df, palette='Set2')
plt.title('Depression Levels')
plt.xlabel('Depression')
plt.ylabel('Number of Students')
plt.show()
```

```python
sns.boxplot(x='gender', y='cgpa', data=df, palette='cool')
plt.title('CGPA by Gender')
plt.show()
```

```python
plt.hist(df['sleep_duration'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Sleep Duration Distribution')
plt.xlabel('Hours')
plt.ylabel('Number of Students')
plt.show()
```

```python
sns.scatterplot(x='work/study_hours', y='cgpa', hue='gender', data=df, palette='husl')
plt.title('Study Hours vs CGPA')
plt.xlabel('Study Hours')
plt.ylabel('CGPA')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get value counts of students by city
city_counts = df['city'].value_counts()

# Optional: Limit to top 5 cities to avoid messy pie chart
top_cities = city_counts[:5]

# Plot the pie chart
plt.figure(figsize=(6,6))
plt.pie(top_cities, labels=top_cities.index, autopct='%1.1f%%', startangle=140, shadow=True, explode=[0.05]*len(top_cities))
plt.title('Top 5 City-wise Student Distribution')
plt.show()

```
